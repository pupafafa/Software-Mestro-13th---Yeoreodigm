```python
import json
import pandas as pd
from pandas.io.json import json_normalize
import requests
api_key = "tvt89yxunjrcdgqy"
df = pd.DataFrame()
request_url = "http://api.visitjeju.net/vsjApi/contents/searchList"
for page_num in range(1,45):
  params = {'apiKey':api_key,'locale':'kr','page':page_num}
  response = requests.get(request_url,params)
  json_file =  json.loads(response.text)
  json_file = json_file['items']
  if page_num == 1:
    df = json_normalize(json_file)
  else:
    tmp_df = json_normalize(json_file)
    df = df.append(tmp_df)
```
