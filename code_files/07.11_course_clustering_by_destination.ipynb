##일정도 클러스터링 항목에 추가

```python
def remove_space(word):
  #print("입력받은 word : ",word)
  return word.replace(" ","")
```

```python
import re
```

```python
x = '무박 3일'
pattern1 = r"\d+박"
pattern2 = r"\무박"
result1 = re.sub(pattern1,repl='',string= x)
print(result1)
result2 = re.sub(pattern2,repl='',string= result1)
print(result2)
pattern3 = r"일"
result3 = re.sub(pattern3,repl='',string = result2)
print(result3)
```

```python
copy_crs = course.copy()
copy_crs['Duration'] = copy_crs['Duration'].apply(remove_space)
copy_crs['Duration'] = copy_crs['Duration'].apply(lambda x: re.sub(pattern1,repl='', string = x))
copy_crs['Duration'] = copy_crs['Duration'].apply(lambda x: re.sub(pattern2,repl='', string = x))
copy_crs['Duration'] = copy_crs['Duration'].apply(lambda x: re.sub(pattern3,repl='', string = x))
#copy_crs['Duration']
```

```python
course['Duration'] = copy_crs['Duration']
```

```python
course.loc[course['Duration']=='Null',['Duration']] =0
```

```python
course.loc[course['Duration']=='Null']
```

```python
grouping_test['Duration'] = course['Duration']
grouping_test.loc[grouping_test['Duration']=='Null']
```

```python
def duration_processing(x):
  #print(x)
  if '시간' in str(x):
    return 0
```
