##식당, 숙박추가

```python
import json
import psycopg2
import pandas as pd
```

```python
tour = pd.read_csv("/content/drive/MyDrive/yeoreodigm/data_files/tourist_attraction_with_rating_sort_by_type.csv",index_col = 0)
```

```python
tour = tour[['contentsid', 'title',  'roadaddress', 'tag','introduction', 'latitude', 'longitude','contentscd.label','phoneno', 'repPhoto.photoid.imgpath', 'repPhoto.photoid.thumbnailpath', '인기점수']]
```

```python
restaurant = tour.loc[tour['contentscd.label']=='음식점']
accomodation = tour.loc[tour['contentscd.label']=='숙박']
```

```python
restaurant['tag_len'] = restaurant['tag'].apply(lambda x: len(x))
accomodation['tag_len'] = accomodation['tag'].apply(lambda x: len(x))
```

```python
restaurant_sort = restaurant.sort_values(by = 'tag_len',ascending=False)
restaurant_sort.head(3)
```

```python
restaurant['tag'].str.get_dummies(sep=',').mean()
```

```python
restaurant['tag'].str.get_dummies(sep=',').mean().sort_values(ascending=False)[:100].index
```

```python
restaurant_top100 = list(restaurant['tag'].str.get_dummies(sep=',').mean().sort_values(ascending=False)[:100].index)
```

###숙박

```python
accomodation_sort = accomodation.sort_values(by = 'tag_len',ascending=False)
accomodation_sort.head(3)
```

```python
accomodation_top40 = accomodation['tag'].str.get_dummies(sep=',').mean().sort_values(ascending=False)[:40].index
```

```python
restaurant['tag'] = restaurant['tag'].apply(lambda x: x.split(','))
accomodation['tag'] = accomodation['tag'].apply(lambda x: x.split(','))
```

```python
restaurant['tag']
```

```python
restaurant['popular_tag'] = restaurant['tag'].apply(lambda x: list(set(x)&set(restaurant_top100)))
accomodation['popular_tag'] = accomodation['tag'].apply(lambda x: list(set(x)&set(accomodation_top40)))
```

```python
restaurant['tag']
```

```python
''.join(['음식','라면'])
```

```python
restaurant['popular_tag'] = restaurant['popular_tag'].apply(lambda x: ','.join(x))
accomodation['popular_tag'] = accomodation['popular_tag'].apply(lambda x: ','.join(x))
```

```python
restaurant['len'] = restaurant['popular_tag'].apply(lambda x: len(x))
accomodation['len'] = accomodation['popular_tag'].apply(lambda x: len(x))
```

```python
restaurant.sort_values(by='len',ascending = False)[['title','popular_tag','len']]
```
