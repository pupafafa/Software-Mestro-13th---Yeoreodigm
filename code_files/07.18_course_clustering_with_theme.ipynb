##테마별 클러스터링을 위한 클러스터링 라이브러리 로드

```python
import sklearn.cluster as cluster
```

##클러스터링 진행

###정규화 이전

```python
kmeans = cluster.KMeans(n_clusters=10,init='k-means++')
tags = ['nature','outdoor','fatigue','sea','walking','exciting','day','culture']

copy_course = course.copy()

kmeans = kmeans.fit(copy_course[tags])
# copy_course['type'] = kmeans.labels_
```

###정규화 (같은 range를 가지는 값이지만, 평균 및 분산이 다르므로 정규화 적용해주기)

```python
from sklearn.preprocessing import MinMaxScaler
data = course[['nature','outdoor','fatigue','sea','walking','exciting','day','culture']]
scaler = MinMaxScaler()
data_scale = scaler.fit_transform(data)


model = cluster.KMeans(n_clusters=12,init='k-means++',random_state=12)
tags = ['nature','outdoor','fatigue','sea','walking','exciting','day','culture']

copy_course = course.copy()

model_result = model.fit(data_scale)
copy_course['cluster'] = model.fit_predict(data_scale)
```

```python
copy_course.sort_values(by='cluster')
```

```python
copy_course.groupby('cluster').mean()
```

```python
copy_course.groupby('cluster').count()['Package_Id']
```
